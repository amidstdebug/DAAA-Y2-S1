```python
import numpy as np
import pandas as pd
from sklearn import datasets
pd.set_option('display.max_rows', 500)
```

# Data Loading and Preparation

```python
iris = datasets.load_iris()
df1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
display(df1)
```

```python
# get the data (X) and target (y)
X,y = iris.data, iris.target
# split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

# Modelling

1. Train the model Features (`X_train`) with Train Labels (`y_train`)
2. Score the model using test features (`X_test`) and test labels (`y_test`)
3. Evaluate the results of model using Confusion Matrix

```python
# get the LogisticRegression estimator
from sklearn.linear_model import LogisticRegression

# training the model
# apply algorithm  to data using fit()
clf = LogisticRegression(solver='newton-cg',multi_class='multinomial')
# train the model with the features (X_train) and labels (y_train)
clf.fit(X_train,y_train)
```

```python
# We can save the trained model clf using pickle (step 4)
import pickle
s = pickle.dumps(clf)
pickle.dump( clf, open( "iris_trained_model.p", "wb" ) )
clf2 = pickle.load( open( "iris_trained_model.p", "rb" ) )

```

```python
# Score the model (step 5)
y_hat = clf.predict(X_test)

# Evaluation of the model (step 6)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_hat)
print(cm)
```
